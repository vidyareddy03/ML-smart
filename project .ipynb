```python
#preprocessing
x=data1.iloc[:,:-1].values
y=data1.iloc[:,-1].values
```

```python
#encode the independent variable state
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_x=LabelEncoder()
x[:,0]=labelencoder_x.fit_transform(x[:,0])
x[:,1]=labelencoder_x.fit_transform(x[:,1])
x[:,4]=labelencoder_x.fit_transform(x[:,4])
x[:,5]=labelencoder_x.fit_transform(x[:,5])
x[:,6]=labelencoder_x.fit_transform(x[:,6])
x[:,9]=labelencoder_x.fit_transform(x[:,9])
print(x)
#onehotencoder=OneHotEncoder(categorical_features=[4,5,9])
#x=onehotencoder.fit_transform(x).toarray()
```

```python
# bcz we have more than 2 levels in each of the columns we need to do onehotencoder
'''ohe=OneHotEncoder(categorical_features=[4])
x=ohe.fit_transform(x).toarray()
ohe=OneHotEncoder(categorical_features=[5])
x=ohe.fit_transform(x).toarray()
ohe=OneHotEncoder(categorical_features=[9])
x=ohe.fit_transform(x).toarray()'''

onehotencoder=OneHotEncoder(categorical_features=[4,5,9])
x = onehotencoder.fit_transform(x).toarray()
```

```python
#apply normalization to rescale the features to a standard range of values.
#normalize the numeric variables from column2 to column31 in the dataframe
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
#splitting the dataset into training set and test set
from sklearn import cross_validation,neighbors
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```
